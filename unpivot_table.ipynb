## How to Unpivot Tables in Python with Pandas
#### By DataDive

```python
import pandas as pd

# a example of wide formated table with sales data for different years and quarters
data = {
    'Year': [2020, 2021],
    'Q1': [200, 220],
    'Q2': [210, 230],
    'Q3': [250, 270],
    'Q4': [240, 260]
}
df= pd.DataFrame(data)
df
```

```python
unpivoted_df = pd.melt(df, id_vars=["Year"])
unpivoted_df
```

```python
unpivoted_df = pd.melt(df, id_vars=["Year"], var_name='Quater',value_name='Sales')
unpivoted_df
```
